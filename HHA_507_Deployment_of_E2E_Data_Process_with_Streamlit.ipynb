### Data cleaning: Removing NaN values from all datasets



```python
df_hospital2 = df_hospital.dropna()
df_hospital2.sample(20)
```

```python
df_inpatient2 = df_inpatient.dropna()
df_inpatient2.sample(20)
```

```python
df_outpatient2 = df_outpatient.dropna()
df_outpatient2.sample(20)
```

### Change 'provider_id' to a string for merging


```python
df_hospital2['provider_id'] = df_hospital2['provider_id'].astype(str)
```

```python
df_inpatient2['provider_id'] = df_inpatient2['provider_id'].astype(str)
```

```python
df_outpatient2['provider_id'] = df_outpatient2['provider_id'].astype(str)
```

### Merge hospital dataframe into outpatient dataframe via left join

```python
df_hospital2_outpatient = df_hospital2.merge(df_hospital2, how = 'left', left_on='provider_id', right_on='provider_id')
df_hospital2_outpatient.sample(20)
```
